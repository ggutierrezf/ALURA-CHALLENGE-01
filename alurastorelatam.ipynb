### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(2)
```

```python
# analisis de datos de la tienda 2
tienda2.head(2)
```

```python
# analisis de datos de la tienda 3
tienda3.head(2)
```

```python
# analisis de datos de la tienda 4
tienda4.head(2)
```

```python
# unir las cuatro tabalas y eliminar el indice y creando un campo que permita saber de que tienda son los datos
tiendas_unidas = pd.concat([tienda.assign(Origen='Tienda 1'), tienda2.assign(Origen='Tienda 2'), tienda3.assign(Origen='Tienda 3'), tienda4.assign(Origen='Tienda 4')]).reset_index(drop=True)
display(tiendas_unidas.head(2))
```

```python
# crear un campo de costo total (precio + envio)
tiendas_unidas['costo total'] = tiendas_unidas['Precio'] + tiendas_unidas['Costo de envío']
display(tiendas_unidas.head(2))
```

```python
# que tipo de datos son?
tiendas_unidas.info()
```

```python
# campo fecha de compra transformado en formato fecha
tiendas_unidas['Fecha de Compra'] = pd.to_datetime(tiendas_unidas['Fecha de Compra'], format='%d/%m/%Y')
display(tiendas_unidas.info())
```

#1. Análisis de facturación



```python
# Analisis de la facturacion por cada tienda y por año
tiendas_unidas['Año'] = tiendas_unidas['Fecha de Compra'].dt.to_period('Y')

tabla_ventas_año_tienda = tiendas_unidas.pivot_table(index='Origen', columns='Año', values='costo total', aggfunc='sum')

tabla_ventas_año_tienda_formatted = tabla_ventas_año_tienda.map(lambda x: f'{x:,.0f}'.replace(',', '.') if pd.notnull(x) else '')

display(tabla_ventas_año_tienda_formatted)
```

```python
# graficando las ventas por año

import matplotlib.pyplot as plt
tabla_ventas_año_tienda.transpose().plot(kind='bar', figsize=(5, 4))

plt.title('Ventas Totales por Año y Tienda')
plt.xlabel('Año')
plt.ylabel('Costo Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# total ventas por Categoria y tienda (usando costo total)

tabla_costo_categoria_tienda = tiendas_unidas.pivot_table(index='Categoría del Producto', columns='Origen', values='costo total', aggfunc='sum')

tabla_costo_categoria_tienda.loc['Total'] = tabla_costo_categoria_tienda.sum()

tabla_costo_categoria_tienda_formatted = tabla_costo_categoria_tienda.map(lambda x: f'{x:,.0f}'.replace(',', '.'))

display(tabla_costo_categoria_tienda_formatted)
```

```python
import matplotlib.pyplot as plt

# Grafico de ventas

tabla_costo_categoria_tienda_for_plot = tabla_costo_categoria_tienda.drop('Total')
tabla_costo_categoria_tienda_for_plot.transpose().plot(kind='bar', figsize=(6, 4))

plt.title('Ventas Totales por Categoría y Tienda (Excluyendo Totales por Tienda)')
plt.xlabel('Origen')
plt.ylabel('Costo Total')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

las tres categorias con mas ventas son : Electronico, Electrodomesticos y Muebles.

# 4. Productos más y menos vendidos

```python
# Ventas de productos (mas y menos vendidos por tienda)

tiendas = tiendas_unidas['Origen'].unique()
for tienda_nombre in tiendas:
    tienda_data = tiendas_unidas[tiendas_unidas['Origen'] == tienda_nombre]
    total_ventas_por_producto_tienda = tienda_data.pivot_table(index='Producto', values='costo total', aggfunc='sum').sort_values(by='costo total', ascending=False)
    total_ventas_por_producto_tienda.plot(kind='bar', figsize=(10, 6))

    plt.title(f'Ventas Totales por Producto en {tienda_nombre} (Ordenado por Volumen de Ventas)')
    plt.xlabel('Producto')
    plt.ylabel('Costo Total')
    plt.xticks(rotation=90, ha='right')
    plt.tight_layout()
    plt.show()
```

los productos mas vendidos por tiendas se mantienen consistentememte en electronica, electrodomesticos, aunque varia los montos y las posiciones relativas entre cada tienda
